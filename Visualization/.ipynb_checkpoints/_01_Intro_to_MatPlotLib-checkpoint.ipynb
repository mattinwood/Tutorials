### Multiple Graphs in a Single Plot

```python
a = plt.plot(flights['passengers'])
c = plt.boxplot(boston['PRICE'])
```

```python
fig = plt.figure()

# Subplot = rows, columns, id
a = fig.add_subplot(1,2,1)
a.plot(flights['passengers'])

c = fig.add_subplot(1,2,2)
c.boxplot(boston['PRICE'])
```

```python
fig = plt.figure()

a = fig.add_subplot(2,1,1)
a.plot(flights['passengers'])

c = fig.add_subplot(2,1,2)
c.boxplot(boston['PRICE'])
```

```python
fig = plt.figure()

a = fig.add_subplot(1,1,1)
a.plot(flights['passengers'])

c = a.twinx()
c.boxplot(boston['PRICE'])
```

```python
fig = plt.figure()

a = fig.add_subplot(1,1,1)
a.plot(flights['passengers'])

c = a.twinx()
c.boxplot(boston['PRICE'])

a.set_xlim([0,150])
```

```python
fig = plt.figure()

a = fig.add_subplot(1,1,1)
a.plot(flights['passengers'])

c = a.twinx().twiny()
c.boxplot(boston['PRICE'])
```
