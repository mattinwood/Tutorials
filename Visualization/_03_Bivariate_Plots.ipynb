```python
# Import packages for plotting and sample data
import matplotlib.pyplot as plt
from matplotlib import legend
import pandas as pd
import seaborn as sns
from sklearn.datasets import load_boston, load_iris
import numpy as np

import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.simplefilter(action='ignore', category=RuntimeWarning)
```

```python
# Import sample data sets
b = load_boston()
boston = pd.DataFrame(data=b.data, columns=b.feature_names)
temp = pd.Series(data=b.target, name="PRICE")
boston = pd.concat([boston, temp], axis=1)
del temp, b

i = load_iris()
iris = pd.DataFrame(data=i.data, columns=i.feature_names)
temp = pd.Series(data=i.target, name="flower")
iris = pd.concat([iris, temp], axis=1)
iris['flower_id'] = iris['flower']
iris.replace({'flower':{0: 'setosa', 1: 'versicolor', 2: 'virginica'}}, inplace=True)
del temp, i

flights = sns.load_dataset('flights')

tips = sns.load_dataset("tips")
```

```python
# The `inline` code allows images to show up within your Notebook, rather than a popup window
%matplotlib inline
```

```python
fig = plt.figure()
a = fig.add_subplot()
a = plt.scatter(iris['sepal length (cm)'], iris['sepal width (cm)'])
```

```python
fig = plt.figure()
a = fig.add_subplot()
a = plt.scatter(iris['sepal length (cm)'], iris['sepal width (cm)'], c=iris['flower_id'], cmap='Set1')
```

```python
fig = plt.figure()
a = fig.add_subplot()
a = plt.scatter(iris['sepal length (cm)'], iris['sepal width (cm)'], c=iris['flower_id'], cmap='Set1', s=50)
```

```python
fig = plt.figure()
a = fig.add_subplot()
a = plt.scatter(tips['total_bill'], tips['tip'], s=50)
```

```python
fig = plt.figure()
a = fig.add_subplot()
tips['sex_id'] = tips['sex'] 
tips.replace({'sex_id':{'Male': 0, 'Female': 1}}, inplace=True)
a = plt.scatter(tips['total_bill'], tips['tip'], c=tips['sex_id'], cmap='Set3', s=50)
```

```python
from sklearn import linear_model as lm
all_x = tips['total_bill'].values.reshape(len(tips), 1)
all_y = tips['tip'].values.reshape(len(tips), 1)

r_all = lm.LinearRegression()
r_all.fit(all_x, all_y)
```

```python
fig = plt.figure()
a = fig.add_subplot()
tips['sex_id'] = tips['sex'] 
tips.replace({'sex_id':{'Male': 0, 'Female': 1}}, inplace=True)
a = plt.scatter(tips['total_bill'], tips['tip'], c=tips['sex_id'], cmap='Set3', s=50)
a = plt.plot(all_x, r_all.predict(all_x))
```
